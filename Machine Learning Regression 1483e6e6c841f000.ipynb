```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(df[['T2']], df[['T6']])
pred = lr.predict(df[['T2']])
```

###### The Coefficient of Determination (R2_score) is  0.44

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

scaler = MinMaxScaler()
X = df.drop(['date', 'Appliances', 'lights'], axis = 1)
target = df['Appliances']
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, target, test_size=0.3, random_state=42)
lr2 = LinearRegression()
lr2.fit(X_train, y_train)
y_pred = lr2.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = round(mean_absolute_error(y_test, y_pred), 2)
print("The Mean Absolute Error is ", mae)
```

```python
rss = round(np.sum(np.square(y_test - y_pred)), 2)
print("The Residual Sum of Square is ", rss)
```

```python
from sklearn.metrics import mean_squared_error
rmse = round(np.sqrt(mean_squared_error(y_test, y_pred)), 3)
print("The Root Mean Squared Error is ", rmse)
```

```python
r2 = round(r2_score(y_test, y_pred), 2)
print("The Coefficient of Determination is ", r2)
```

```python
# Obtaining the feature weights
weight_df = pd.DataFrame(lr2.coef_, index = X.columns, columns = ['Feature_Weights'])
weight_df
```

```python
weight_df.sort_values('Feature_Weights', ascending = True)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha= 0.4)
ridge_reg.fit(X_train, y_train)
ridge_pred = ridge_reg.predict(X_test)
rmse2 = round(np.sqrt(mean_squared_error(y_test, ridge_pred)), 3)
print("The Root Mean Squared Error when trained using Ridge Regression is ", rmse2)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha= 0.001)
lasso_reg.fit(X_train, y_train)
weight_df2 = pd.DataFrame(lasso_reg.coef_, index = X.columns, columns = ['Feature_Weights'])
weight_df2
```

```python
lasso_pred = lasso_reg.predict(X_test)
rmse3 = round(np.sqrt(mean_squared_error(y_test, lasso_pred)), 3)
print("The Root Mean Squared Error when trained using Lasso Regression is ", rmse3)
```
